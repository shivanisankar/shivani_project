```python
df1 = pd.read_csv("zendesk_tickets_combined_3.csv")
print(df1)
```

```python
#changing user id to a string, instead of a number, so i can concat
df2 = df1.astype({"Ticket Requesters User ID": str})
print(df2)
```

```python
#creating concat key
df2['concat_key'] = df2['Ticket Requesters User ID'] + df2['Ticket Requesters Name'] + df2['Ticket Requesters Client Name']
print(df2)
```

```python
#creating pivot table
df3 = pd.pivot_table(df2,index=["Ticket Requesters User ID","Ticket Requesters Name","Ticket Requesters Client Name",
                          "concat_key"],columns=["Zendesk Ticket Tags EPD Component","Zendesk Tickets Reason", 
                        "Zendesk Tickets Historical Severity"])
print(df3)
```

```python
#printing pivot table again -- easier to see
pd.pivot_table(df2,index=["Ticket Requesters User ID","Ticket Requesters Name","Ticket Requesters Client Name",
                          "concat_key"],columns=["Zendesk Ticket Tags EPD Component","Zendesk Tickets Reason", 
                        "Zendesk Tickets Historical Severity"])
```
